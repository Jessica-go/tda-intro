# Persistent Homology on a Circle (S¹)

In this notebook, I generate a point cloud sampled from a circle and compute its persistent homology using Ripser.

Goals:
- Understand 0-dimensional homology (connected components)
- Understand 1-dimensional homology (the loop of the circle)
- Visualize persistence diagrams and barcodes


```python
import numpy as np
import matplotlib.pyplot as plt
from ripser import ripser
from persim import plot_diagrams

```

## 1. Generate a point cloud on a circle


```python
# Number of points on the circle
n_points = 100

# Angles from 0 to 2π
theta = np.linspace(0, 2 * np.pi, n_points, endpoint=False)

# Coordinates on the unit circle
X = np.vstack((np.cos(theta), np.sin(theta))).T

X[:5]
```

```python
plt.figure(figsize=(5,5))
plt.scatter(X[:, 0], X[:, 1])
plt.gca().set_aspect("equal")
plt.title("Point cloud sampled from a circle")
plt.show()
```

## 2. Compute persistent homology with Ripser


```python
result = ripser(X)
diagrams = result["dgms"]
diagrams
```
